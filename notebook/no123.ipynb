```python
def show_pareto(df, column_name):
    column_value_counts = df.value_counts(column_name)
    column_value_counts = df.value_counts(column_name).reset_index()
    total_count = len(df)
    column_value_counts["cumsum"] = (
        column_value_counts["count"].cumsum() * 100 / total_count
    )
    column_value_counts_80 = column_value_counts[column_value_counts["cumsum"] <= 80]
    fig = go.Figure()
    fig.add_trace(
        go.Bar(
            x=column_value_counts_80[column_name],
            y=column_value_counts_80["count"],
            name="count",
            marker=dict(color="steelblue"),
        )
    )

    fig.add_trace(
        go.Scatter(
            x=column_value_counts_80[column_name],
            y=column_value_counts_80["cumsum"],
            name="Cumulative Percentage",
            yaxis="y2",
            mode="lines",
        )
    )
    fig.update_layout(
        title="Pareto Chart",
        yaxis=dict(title="Count"),
        yaxis2=dict(
            title="Cumulative %",
            overlaying="y",
            side="right",
            range=[0, 100],  # percentage scale
        ),
        xaxis=dict(title=column_name),
    )

    return fig, column_value_counts_80
```

```python
fig, column_value_counts_80 = show_pareto(df, "country")
fig.show()
```

### No 2 
From the csv file size lets just parse this using the combination of DuckDB+Pandas and compare it with Basic Pandas.

DuckDB is used to load the data schema and basic filtering and groupby processing to lower RAM usage and lower processing time.
Pandas is still used after words for easy data analytics after we got smaller subset of the data.
This is possible because the size of data is not that big (only ~300MB csv) if the data is >10GB csv then we might need to go full DuckDB

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import psutil, os, time

start_time = time.time()
pid = os.getpid()
process = psutil.Process(pid)
start_mem = process.memory_info().rss / 1024**2
print(f"Starting RAM Usage: {start_mem:.2f} MB")

data_path = "data/customers-2000000.csv"
df = pd.read_csv(data_path)

print(df.columns)

current_mem = process.memory_info().rss / 1024**2
print(f"Loading Data Schema RAM Usage: {(current_mem - start_mem):.2f} MB")

df = df[["Customer Id", "Country", "Subscription Date"]]

df_country_value_counts = df["Country"].value_counts().reset_index()
current_mem = process.memory_info().rss / 1024**2
print(f"Country Profiling RAM Usage: {(current_mem - start_mem):.2f} MB")


df["Subscription Date"] = pd.to_datetime(df["Subscription Date"])
df["year"] = df["Subscription Date"].dt.year
df["month"] = df["Subscription Date"].dt.month


trend_df = df.groupby(["year", "month"])["Customer Id"].agg("count").reset_index()
trend_df["dt"] = (
    trend_df["year"].astype(str) + "-" + trend_df["month"].astype(str).str.zfill(2)
)
trend_df["moving_avg"] = trend_df["Customer Id"].rolling(window=6, min_periods=1).mean()

mean_val = trend_df["Customer Id"].mean()
std_val = trend_df["Customer Id"].std()
upper_bound = mean_val + 3 * std_val
lower_bound = mean_val - 3 * std_val

current_mem = process.memory_info().rss / 1024**2
print(f"Subscription Rate Profiling RAM Usage: {(current_mem - start_mem):.2f} MB")

print(f"Time taken : {time.time() - start_time} s")

fig = px.box(df_country_value_counts, y="count")
fig.show()

fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=trend_df["dt"],
        y=trend_df["Customer Id"],
        name="customer",
        mode="lines+markers",
    )
)
fig.add_hline(
    y=mean_val,
    line_dash="dot",
    line_color="red",
    annotation_text=f"Mean = {mean_val:.2f}",
    annotation_position="top left",
)
fig.add_trace(
    go.Scatter(
        x=trend_df["dt"],
        y=trend_df["moving_avg"],
        mode="lines+markers",
        name="Moving Avg (6M)",
        line=dict(color="orange"),
    )
)

fig.add_hline(
    y=upper_bound,
    line_dash="dash",
    line_color="green",
    annotation_text=f"+3Ïƒ = {upper_bound:.2f}",
    annotation_position="bottom left",
)

fig.add_hline(
    y=lower_bound,
    line_dash="dash",
    line_color="green",
    annotation_text=f"-3Ïƒ = {lower_bound:.2f}",
    annotation_position="top left",
)

fig.update_layout(
    title="Value with Moving Average & Mean",
    xaxis_title="Year-Month",
    yaxis_title="Value",
)


fig.show()
```

```python
import duckdb
import plotly.graph_objects as go
import plotly.express as px
import psutil, os, time
import pandas as pd

start_time = time.time()
pid = os.getpid()
process = psutil.Process(pid)
start_mem = process.memory_info().rss / 1024**2
print(f"Starting RAM Usage: {start_mem:.2f} MB")

data_path = "data/customers-2000000.csv"

con = duckdb.connect()

schema = con.execute(f"""
    DESCRIBE SELECT * FROM read_csv_auto('{data_path}') LIMIT 0
""").fetchdf()

print(schema)

current_mem = process.memory_info().rss / 1024**2
print(f"Loading Data Schema RAM Usage: {(current_mem - start_mem):.2f} MB")

df_country_value_counts = con.execute(f"""
    SELECT Country, COUNT("Customer Id") as count
    FROM read_csv_auto('{data_path}', SAMPLE_SIZE=1)
    GROUP BY Country       
""").fetch_df()

current_mem = process.memory_info().rss / 1024**2
print(f"Country Profiling RAM Usage: {(current_mem - start_mem):.2f} MB")

trend_df = con.execute(f"""
        SELECT 
            CAST(strftime('%Y', "Subscription Date") AS INT) AS year,
            CAST(strftime('%m', "Subscription Date") AS INT) AS month,
            COUNT("Customer Id") AS count
        FROM read_csv_auto('{data_path}', SAMPLE_SIZE=1)
        GROUP BY year, month
        ORDER BY year, month
""").fetch_df()

trend_df["dt"] = (
    trend_df["year"].astype(str) + "-" + trend_df["month"].astype(str).str.zfill(2)
)
trend_df["moving_avg"] = trend_df["count"].rolling(window=6, min_periods=1).mean()

mean_val = trend_df["count"].mean()
std_val = trend_df["count"].std()
upper_bound = mean_val + 3 * std_val
lower_bound = mean_val - 3 * std_val

current_mem = process.memory_info().rss / 1024**2
print(f"Subscription Rate Profiling RAM Usage: {(current_mem - start_mem):.2f} MB")

print(f"Time taken : {time.time() - start_time} s")

fig = px.box(df_country_value_counts, y="count")
fig.show()


fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=trend_df["dt"],
        y=trend_df["count"],
        name="customer",
        mode="lines+markers",
    )
)
fig.add_hline(
    y=mean_val,
    line_dash="dot",
    line_color="red",
    annotation_text=f"Mean = {mean_val:.2f}",
    annotation_position="top left",
)
fig.add_trace(
    go.Scatter(
        x=trend_df["dt"],
        y=trend_df["moving_avg"],
        mode="lines+markers",
        name="Moving Avg (6M)",
        line=dict(color="orange"),
    )
)

fig.add_hline(
    y=upper_bound,
    line_dash="dash",
    line_color="green",
    annotation_text=f"+3Ïƒ = {upper_bound:.2f}",
    annotation_position="bottom left",
)

fig.add_hline(
    y=lower_bound,
    line_dash="dash",
    line_color="green",
    annotation_text=f"-3Ïƒ = {lower_bound:.2f}",
    annotation_position="top left",
)

fig.update_layout(
    title="Value with Moving Average & Mean",
    xaxis_title="Year-Month",
    yaxis_title="Value",
)


fig.show()
```
